## 컬럼추가히기

```python
train["DatasetType"]="Titanic"

# 개수만 맞으면 뭐든지 들어간다
train["Id"]=range(0,891)
train
```

```python
#응용
train["FamilySize"]=train["SibSp"]+train["Parch"]+1 #나자신 포함
#train.head()

train[["SibSp","Parch",'FamilySize']].head()
```

```python
train["Embarked_C"]= train["Embarked"] =="C" # true, false 값을 새로만든 컬럼에 넣는다
train["Embarked_S"]= train["Embarked"] =="S"
train["Embarked_Q"]= train["Embarked"] =="Q"

train[["Embarked","Embarked_C","Embarked_S","Embarked_Q"]].head() # one_hot encoding이라고 한다.
#train.head()
```

```python
train["Fare_low"]= train["Fare"] < 20 
train["Fare_medium"]= (train["Fare"] >= 20) & (train["Fare"] < 40) 
train["Fare_high"]= train["Fare"] >= 40 

train[["Fare","Fare_low","Fare_medium","Fare_high" ]].head()
```

```python
train['Age'].isnull()
train[train['Age'].isnull()].head() 
train.loc[train['Age'].isnull(),"Age"] # train에서 row가 null이면서 컬럼이 "AGE"인 데이터만 가져온다
train.loc[train['Age'].isnull(),"Age"] = train["Age"].mean()
train.head()
```
